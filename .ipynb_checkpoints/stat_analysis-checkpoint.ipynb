## Analysis of cross-regulation predictions

```python
#packages
import pandas as pd
import seaborn as sns
```

```python
cross_targets_1 = pd.read_table('./data/outputs/', header=None)
cross_targets.columns = ['factor', 'cell line', 'exon', 'target', 'deltaPSI', 'deltaPSIc', 'z', 'p', 'q', 'KDFC']
NMD_KD = pd.read_table('./data/input_data/upf1xrn1_deltaPSI.tsv')
```

```python
NMD_KD.head()
```

```python
nmd_kd = NMD_KD[['id', 'gene', 'deltaPSI.UPF1', 'deltaPSI.SMG6']]
nmd_kd = nmd_kd.rename(columns={'id':'exon'})
nmd_kd.head()
```

```python
#nmd and rbps KD deltaPSIs
two_kds = pd.merge(nmd_kd, cross_targets, on='exon')
two_kds.head()
```

### SRSF7

```python
# gcn_data
srsf7_gcn = pd.read_table('./srsf7.tsv')
# cross_targets

cross_srsf7 = cross_targets[cross_targets['factor'] == 'SRSF7']
cross_srsf7
```

```python
srsf7_cross_targets = list(cross_srsf7['target'])
```

```python
srsf7_gcn[srsf7_gcn['target'] == 'PRPF39']
```

```python
srsf7_gcn_list = list(srsf7_gcn['target'])
list(set(srsf7_cross_targets) & set(srsf7_gcn_list))
```

```python
#nmd and rbps KD deltaPSIs
two_kds = pd.merge(nmd_kd, cross_targets, on='exon')
two_kds[two_kds['factor'] == 'SRSF7']
```
