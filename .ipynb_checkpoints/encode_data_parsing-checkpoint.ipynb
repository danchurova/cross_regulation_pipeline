### ENCODE DATA

```python
# packages
import pandas as pd
```

```python
# import GTEx data
data = pd.read_csv('m2.tsv', sep='\t')
```

```python
data.head()
```

```python
colnames = ['gene']
for i in data.columns[1:]:
    i = i.split('.')[0]
    colnames.append(i)
```

```python
data.columns = colnames
```

```python
# 740 samples
len(data.columns) 
```

```python
#gene_id - import and clean
gene_id = pd.read_table('./id_gene_name_list.txt', header=None)
gene_id.columns = ['ID', 'gene']
gene_id['ID'] = gene_id['ID'].str.replace(";","")
gene_id['gene'] = gene_id['gene'].str.replace(";","")
gene_id['ID'] = gene_id['ID'].str.split(".").str[0]
gene_id.set_index('ID',inplace=True)
gene_id_dict = gene_id.to_dict()['gene']
```

```python
data['gene'] = data['gene'].map(gene_id_dict) 
```
