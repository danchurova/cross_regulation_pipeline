### Parsing Kulakovsky's data

```python
# packages
import pandas as pd
```

```python
# data
srsf7_K562 = pd.read_csv('./Result_genes_ENS/SRSF7_K562.txt', sep='\t')
srsf7_hepG2 = pd.read_csv('./Result_genes_ENS/SRSF9_HepG2.txt', sep='\t')

#IDs to names
ids_gene_names = pd.read_csv('./id_gene_name_list.txt', sep='\t', header=None)
```

```python
srsf7_hepG2.head()
```

```python
ids_gene_names.columns = ['ID', 'Gene name']
ids_gene_names.head()
```

```python
ids_gene_names["ID"]= ids_gene_names["ID"].str.split(".", n = 1, expand = True) 
ids_gene_names["Gene name"]= ids_gene_names["Gene name"].str.split(";", n = 1, expand = True) 
```

```python
ids_names_dict = ids_gene_names.set_index('ID')['Gene name'].to_dict()
```

```python
srsf7_hepG2 = srsf7_hepG2.replace({"Gene_ID": ids_names_dict})
srsf7_K562 = srsf7_K562.replace({"Gene_ID": ids_names_dict})
```

```python
#srsf7_K562.sort_values(by='FDR',ascending=False)
srsf7_targets_merged = pd.merge(srsf7_hepG2, srsf7_K562, on='Gene_ID', how='inner')
srsf7_targets_merged.columns = ['Gene_name', 'pval_HepG2', 'FDR_HepG2', 'logFC_HepG2', 'z_score_HepG2', \
                         'logCPM_HepG2', 'pval_K562', 'FDR_K562', 'logFC_K562', 'z_score_K562','logCPM_K562']
srsf7_targets_merged
```

```python
srsf7_targets_merged.to_csv('merged_eCLIP_targets.tsv', sep = '\t')
```
