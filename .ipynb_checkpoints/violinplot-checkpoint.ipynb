```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('/Users/danchurova/Downloads/trnas_lenghts.csv')
```

```python
transposed = data.T
transposed.columns = list(transposed.iloc[0,])
transposed = transposed.iloc[1:]
transposed
```

```python
trans_group = transposed.melt(var_name='aminoacids', value_name='vals')
trans_group.vals = trans_group.vals.astype(float) # тут исправляю ошибку, ему не нравится, что тип данных object
trans_group
```

```python
ax = sns.violinplot(x="aminoacids", y="vals", data=trans_group)
```
